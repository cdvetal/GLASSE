```python
import pandas as pd
import csv
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
import matplotlib.font_manager as fm
```

```python
gen = []
for i in range(0, 400):
  gen.append(i)
```

```python
for number in range(0,10):
  # path for stats 
  # path = f'./stat_digit_{number}'
  path = '.'
  
  adv_transport = f'{path}/adv_transport_{number}.csv'
  f_adv = open(adv_transport, 'w')
  writer_adv = csv.writer(f_adv)

  inc_classif = f'{path}/inc_classif_{number}.csv'
  f_i = open(inc_classif, 'w')
  writer_i = csv.writer(f_i)

  fit_big = f'{path}/fit_big_{number}.csv'
  f_b = open(fit_big, 'w')
  writer_b = csv.writer(f_b)

  writer_adv.writerow(gen)
  writer_i.writerow(gen)
  writer_b.writerow(gen)

  for run in range(1, 11):
    df = pd.read_csv(f"{path}/Arquivo_digit_{number}_run_{run}.csv") 
    num_adv = df["numero de adversariais"]
    writer_adv.writerow(num_adv)

    num_i = df['numero de classificações erradas']
    writer_i.writerow(num_i)

    df_f = pd.read_csv(f"{path}/Fit_digit_{number}_run_{run}.csv")
    biggest_fit = df_f['fitness biggest']
    writer_b.writerow(biggest_fit)

  f_adv.close()
  f_i.close()
  f_b.close()
```

```python
sns.set(style="white", rc={'figure.figsize':(12, 8), 'pdf.fonttype': 42, 'ps.fonttype': 42})

plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman'] + plt.rcParams['font.serif']
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42
```

boxplot -> one number only, must be few generations

```python
xl = 'generation'
yl = 'quantity of adversarials'

df = pd.read_csv(f'adv_transport_{number}.csv')
gra = sns.boxplot(data=df, width=0.3, medianprops={"color": "red"})
gra.set_xlabel(xl,fontdict= {'fontsize': 25})
gra.set_ylabel(yl,fontdict= {'fontsize': 25})
# Set tick font size
for label in (gra.get_xticklabels() + gra.get_yticklabels()):
	label.set_fontsize(18)
plt.xticks(rotation=45)
plt.show()
```

average (across runs) number of adversarials

```python
fig, ax = plt.subplots(5,2, sharex=True, sharey=True)
col = 0
lin = 0
xl = 'generations'
yl = 'mean number of adversarials'

for number in range(0,10):
  df = pd.read_csv(f'adv_transport_{number}.csv')
  mean = df.mean()
  if number < 5:
    col = 0
    lin = number
  else:
    col = 1
    lin = number - 5
  ax[lin, col].plot(gen, mean)
  ax[lin, col].set_title(number, fontdict= {'fontsize': 10})


ax[4,0].set(xlabel=xl)
ax[4,0].set_xlabel(xl,fontdict= {'fontsize': 18})
ax[4,1].set(xlabel=xl)
ax[4,1].set_xlabel(xl,fontdict= {'fontsize': 18})
ax[2,0].set(ylabel=yl)
ax[2,0].set_ylabel(yl,fontdict= {'fontsize': 18})
```
